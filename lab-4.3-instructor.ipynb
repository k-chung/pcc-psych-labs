<font color="brown" size="4">üî¢ Accessing Data: What is an Index?</font>

An **index** is the number used to data from an ordered structure (like a column of a table, or an array). You'll see indexes used everywhere in Python.

The most important rule in Python is that **indexing always starts at 0** (the first element).

| Structure | Syntax | Example |
| :--- | :--- | :--- |
| **List/Array** | `list_name[position]` | `my_list[0]` retrieves the **first** item. |

**Question 1.2:** Fill in the following code: what's the right index  to access the 2nd column of the table?

```python
observed = PI_students.column(1)
# observed = PI_students.column(...)

print("observed column as an array:", observed)
print(type(observed))
```

**Question 1.3:** Get the total number of students by using `sum()` on `observed`.

```python
PI_students_N = sum(observed)
# PI_students_N = sum(...)

print("Total number of students:", PI_students_N)
```

**Question 1.4:** Calculate the degrees of freedom (Use this formula: Ncategories ‚Äì 1)
*Note: This is a Chi-square test for goodness of fit since there is just 1 nominal variable.

```python
PI_students_Ncategories = 5
df = PI_students_Ncategories - 1
# PI_students_Ncategories = ...
# df = PI_students_Ncategories - ...

print("Degrees of freedom:", df)
```

**Question 1.5:** Calculate the expected frequency (Hint: Use the formula: N / Ncategories).


```python
expected_freq = PI_students_N / PI_students_Ncategories
# expected_freq = ... / ...

print(f"Expected Frequencies: {expected_freq}")
```

<font color="brown" size="4">‚úèÔ∏è Manually calculate chi square</font>

We _could_ just use Python to calculate chi square for us, but it's better for your understanding if you calculate it manually, for now. These steps are very similar to what you'd do in Excel.

**Question 1.6:** Get the array of observed frequencies. It's the second column in the table. What index would you use to access it?

```python
observed_freq = PI_students.column(1)
# observed_freq = PI_students.column(...)

print(f"Observed Frequencies: {observed_freq}")
```

**Question 1.7:** Define a new array that is the difference between observed and expected. You'll be using `observed_freq` (which you just defined), and the previous `expected_freq`.

Hint: when you subtract a singular number from an array, it returns a new array that applies that subtraction to _all_ the numbers. For example: [1, 2, 3] - 1 = [0, 1, 2]

```python
diff_arr = observed_freq - expected_freq
# diff_arr = ... - ...

print("diff_arr:", diff_arr)
```

**Question 1.8:** Define a new array by squaring `diff_arr`.

```python
diff_arr_squared = diff_arr**2
# diff_arr_squared = ...**2

print("diff_arr_squared:", diff_arr_squared)
```

**Question 1.9:** Create a new array by dividing `diff_arr_squared` by `expected_freq`.

```python
division_array = diff_arr_squared / expected_freq
# division_array = ... / ...

print('division_array:', division_array)
```

**Question 1.10:** Finally, sum up all the terms in `division_array`. This is your chi square value!

```python
chi_square = sum(division_array)
# chi_square = sum(...)

print('Chi-square (X¬≤):', chi_square)
```

**Question 1.11:** Calculating p-value manually is outside the scope of this lab, so instead we'll use a shortcut: the `scipy.stats.chisquare` function!

Plug in two arguments: the array `observed` and the number `expected_freq`, which you both previously defined.

```python
# Use these variables (printed for your convenience)
print("observed:", observed)
print("expected frequency:", expected_freq)

chi_square_statistic, p_value = chisquare(observed, expected_freq)
# chi_square_statistic, p_value = chisquare(..., ...)

print("p-value:", p_value)
```
